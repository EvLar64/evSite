```python
import nltk
import nltk.corpus
from nltk.book import *
# The next line lets us do GET requests from remote URLs on the web:
from urllib import request
# The following import lines are for plotting interactive visualizations in Python
import matplotlib
import matplotlib.pyplot as plt
import tk
import nltk, re, pprint
from nltk import word_tokenize

# Reading in 210 project files
with open("ferris.txt",'r') as file:
    ferris=file.read()
with open("breakfast.txt",'r') as file:
    breakfast=file.read()
with open("sixteen.txt",'r') as file:
    sixteen=file.read()

# Tokenizing project files
bueller = word_tokenize(ferris)
club = word_tokenize(breakfast)
candles = word_tokenize(sixteen)

# Testing tokenization
print("Example line showing succesful tokenization:")
print(bueller[:17])

# Making dispersion plots with common words for each screenplay
words = ["the", "and", "but", "for"]
nltk.draw.dispersion_plot(bueller, words)
plt.title("Ferris Bueller - Lexical Dispersion Plot")
plt.show()

nltk.draw.dispersion_plot(club, words)
plt.title("Breakfast Club - Lexical Dispersion Plot")
plt.show()

nltk.draw.dispersion_plot(candles, words)
plt.title("Sixteen Candles - Lexical Dispersion Plot")
plt.show()

# Attempted to use various word functions including 'similar' and 'common_contexts', 
# but on each attempt I was met with an error that the "list"/"str" does not posses a "similar" attribute. 
# I tried with both versions of each script and was met with the same error (only difference being that "str" vs. "list" thing)
# I'm not sure if it's an install error or import failure on my part, but in the current state I'm not able to access those various word functions.

#bueller.common_contexts(["Ferris","Bueller"])
#candles.similar("birthday")
```
